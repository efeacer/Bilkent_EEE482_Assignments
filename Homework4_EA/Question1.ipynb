```python
# Necessary imports
import numpy as np
import matplotlib.pyplot as plt
import h5py # to be able to use v 7.3 .mat files in the Python 
from sklearn.decomposition import PCA 
```

```python
with h5py.File('hw4_data1.mat', 'r') as file:
    data_keys = list(file.keys())

data = dict()
with h5py.File('hw4_data1.mat', 'r') as file:
    for key in data_keys:
        data[key] = np.array(file[key])
        print('Shape of the data associated with %s:' % key,
              np.shape(data[key]), '\n')
```

```python
# QUESTION 1
print('QUESTION 1')
```

```python
faces = data['faces'].T
```

```python
# A sample stimuli
figure_num = 1
plt.figure(figure_num)
figure_num += 1
plt.imshow(faces[7].reshape(32, 32).T, cmap=plt.cm.bone)
plt.title('Sample Stimuli')
plt.show(block=False)
```

```python
# PART A
print('PART A')
```

```python
pca = PCA(100) 
pca.fit(faces) 
```

```python
plt.figure(figure_num)
figure_num += 1
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('PC')
plt.ylabel('Explained Variance')
plt.title('Proportion of Variance Explained by Each Individual PC')
plt.grid()
plt.show(block=False)
```

```python
plt.figure(figure_num)
figure_num += 1
fig, axes = plt.subplots(5, 5, figsize=(10,10),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i, ax in enumerate(axes.flat):
     ax.imshow(pca.components_[i].reshape(32, 32).T, cmap=plt.cm.bone)
fig.suptitle('PCs', fontsize=20)
plt.plot(block=False)
```

```python
cov= pca.get_covariance()
eigvals, eigvecs = np.linalg.eig(cov)
```

```python
NUM_FEATURES = pca.components_.shape[1]
```

```python
reconstruction10 = np.zeros(NUM_FEATURES)
for i in range(10):
    reconstruction10 += pca.components_[i] * eigvals[i].real
```

```python
figure_num = 1
plt.figure(figure_num)
figure_num += 1
plt.imshow(reconstruction10.reshape(32, 32).T, cmap=plt.cm.bone)
plt.title('Sample Stimuli')
plt.show(block=False)
```
